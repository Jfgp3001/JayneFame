leap year

```python
year= eval(input("Please input the year"))

if ((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is leap year".format(year))
else:
  print("{0} is not a leap year".format(year))
```

For Loop

```python
fruits = ["mangosteen", "kiwi", "grapes"]
for x in fruits:
  print(x)
```

```python
for x in "mangosteen":
  print(x)
```

```python
for x in range(20):
  print(x)
```

```python
fruits = ["mangosteen", "kiwi", "grapes"]
for x in fruits:
  print(x)
  if x == "mangosteen":
    break
```

```python
for x in range(20):
  print(x)
  if x == 5:
    break
```

```python
for x in range(20):
    if x == 5:
        continue
    print(x)
```

```python
for x in range(22):
   if x%6 == 0:
    continue

   
   print(x)
```
